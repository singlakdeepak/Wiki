```python
N = np.array ([ 53,69,69,59,70,54,59,80,29,30])
def eqn1(w,h,N):
    return -18*w -13*h + 7*N
def eqn2(w,h,N):
    return -42*w + 19*h +17*N
def eqn3(w,h,N):
    return 72*w + 10*h -98*N

w = np.linspace(0,120,500)
h = np.arange(0,60,5)

for hval in h:
    plt.plot(w, eqn1(w, hval,53))
plt.show()
```

```python
import pandas as pd

gl = pd.read_csv('../Sheet1.csv')
gl.head()
```

```python
Car_van_data = gl.loc[0:1,'Test1':'test10'].values
Bus_data = gl.loc[2:3,'Test1':'test10'].values
Human_data = gl.loc[4:5,'Test1':'test10'].values
Bike_data = gl.loc[6:7,'Test1':'test10'].values
Truck_data = gl.loc[8:9,'Test1':'test10'].values
```

```python
def plot3dres(data, results,N, eqn2use=1):
    fig = plt.figure()

    # Add an axes
    ax = fig.add_subplot(111,projection='3d')
    ww, hh = np.meshgrid(range(120), range(60))
    # plot the surface
    if eqn2use ==1:
        ax.plot_surface(ww, hh, eqn1(ww,hh,N), alpha=0.2)
    elif eqn2use ==2:
        ax.plot_surface(ww, hh, eqn2(ww,hh,N), alpha=0.2)
    else:
        ax.plot_surface(ww, hh, eqn3(ww,hh,N), alpha=0.2)
    

    # point2 = np.array([10, 50, 50])

    # and plot the point 
    for counter, value in enumerate(results):
        if value>0:
            ax.scatter(data[0][counter],data[1][counter],value,  color='red')
        else:
            ax.scatter(data[0][counter],data[1][counter],value,  color='green')
        

```

```python
%matplotlib inline
# Create the figure
Car_results = eqn1(Car_van_data[0],Car_van_data[1],N)
Bus_results = eqn1(Bus_data[0],Bus_data[1],N)
Human_results = eqn1(Human_data[0],Human_data[1],N)
Bike_results = eqn1(Bike_data[0],Bike_data[1],N)
Truck_results = eqn1(Truck_data[0],Truck_data[1],N)
```

```python

# test1 to test9 gets separated but test10 is on the same side for Human class
plot3dres(Truck_data,Truck_results,53, eqn2use =1) 
```

```python
Car_results2 = eqn2(Car_van_data[0],Car_van_data[1],N)
Bus_results2 = eqn2(Bus_data[0],Bus_data[1],N)
Human_results2 = eqn2(Human_data[0],Human_data[1],N)
Bike_results2 = eqn2(Bike_data[0],Bike_data[1],N)
Truck_results2 = eqn2(Truck_data[0],Truck_data[1],N)
```

```python
# human class for all the test sets is above the plane
# Bike class for all the test sets is above the plane
plot3dres(Truck_data,Truck_results2,53,eqn2use=2)
```

```python
Car_results3 = eqn3(Car_van_data[0],Car_van_data[1],N)
Bus_results3 = eqn3(Bus_data[0],Bus_data[1],N)
Human_results3 = eqn3(Human_data[0],Human_data[1],N)
Bike_results3 = eqn3(Bike_data[0],Bike_data[1],N)
Truck_results3 = eqn3(Truck_data[0],Truck_data[1],N)
```

```python
# Bus data for all the test sets is above the plane
plot3dres(Truck_data,Truck_results3,53,eqn2use=3)
```
